# Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score

# Data Preparation
X = heart_data.drop("target", axis=1)
y = heart_data["target"]
```

```python
# Divide the dataset in 70:30 ratio
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Build the Logistic Regression model
log_reg_model = LogisticRegression(random_state=42)
log_reg_model.fit(X_train, y_train)
```

```python
# Predict the values on the test set
y_pred_log_reg = log_reg_model.predict(X_test)

# Build the confusion matrix and get the accuracy score
conf_matrix_log_reg = confusion_matrix(y_test, y_pred_log_reg)
accuracy_log_reg = accuracy_score(y_test, y_pred_log_reg)

print("Confusion Matrix (Logistic Regression):")
print(conf_matrix_log_reg)
print("Accuracy Score (Logistic Regression):", accuracy_log_reg)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
# Build the Decision Tree model
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

# Predict the values on the test set
y_pred_dt = dt_model.predict(X_test)

# Build the confusion matrix and calculate the accuracy
conf_matrix_dt = confusion_matrix(y_test, y_pred_dt)
accuracy_dt = accuracy_score(y_test, y_pred_dt)

print("Confusion Matrix (Decision Tree):")
print(conf_matrix_dt)
print("Accuracy Score (Decision Tree):", accuracy_dt)

```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

# Build the Random Forest model
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# Predict the values on the test set
y_pred_rf = rf_model.predict(X_test)

# Build the confusion matrix and calculate the accuracy
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
accuracy_rf = accuracy_score(y_test, y_pred_rf)

print("Confusion Matrix (Random Forest):")
print(conf_matrix_rf)
print("Accuracy Score (Random Forest):", accuracy_rf)
```

# Select the Best Model

```python
from sklearn.metrics import classification_report

# Print the confusion matrix of all classifiers
print("Confusion Matrix (Logistic Regression):")
print(conf_matrix_log_reg)
print("Confusion Matrix (Decision Tree):")
print(conf_matrix_dt)
print("Confusion Matrix (Random Forest):")
print(conf_matrix_rf)
```

```python
# Print the classification report of all classifiers
print("Classification Report (Logistic Regression):")
print(classification_report(y_test, y_pred_log_reg))
print("Classification Report (Decision Tree):")
print(classification_report(y_test, y_pred_dt))
print("Classification Report (Random Forest):")
print(classification_report(y_test, y_pred_rf))
```

```python
# Calculate Recall, Precision, and F1 score of all the models
from sklearn.metrics import recall_score, precision_score, f1_score

recall_log_reg = recall_score(y_test, y_pred_log_reg)
precision_log_reg = precision_score(y_test, y_pred_log_reg)
f1_score_log_reg = f1_score(y_test, y_pred_log_reg)

recall_dt = recall_score(y_test, y_pred_dt)
precision_dt = precision_score(y_test, y_pred_dt)
f1_score_dt = f1_score(y_test, y_pred_dt)

recall_rf = recall_score(y_test, y_pred_rf)
precision_rf = precision_score(y_test, y_pred_rf)
f1_score_rf = f1_score(y_test, y_pred_rf)

print("Logistic Regression - Recall: {}, Precision: {}, F1-score: {}".format(recall_log_reg, precision_log_reg, f1_score_log_reg))
print("Decision Tree - Recall: {}, Precision: {}, F1-score: {}".format(recall_dt, precision_dt, f1_score_dt))
print("Random Forest - Recall: {}, Precision: {}, F1-score: {}".format(recall_rf, precision_rf, f1_score_rf))
```

```python
# Visualize confusion matrix using heatmaps
plt.figure(figsize=(10, 6))
sns.heatmap(conf_matrix_log_reg, annot=True, cmap="Blues", fmt="d", cbar=False)
plt.title("Confusion Matrix - Logistic Regression")
plt.show()

plt.figure(figsize=(10, 6))
sns.heatmap(conf_matrix_dt, annot=True, cmap="Greens", fmt="d", cbar=False)
plt.title("Confusion Matrix - Decision Tree")
plt.show()

plt.figure(figsize=(10, 6))
sns.heatmap(conf_matrix_rf, annot=True, cmap="Oranges", fmt="d", cbar=False)
plt.title("Confusion Matrix - Random Forest")
plt.show()
```

```python
# Select the best model based on the best accuracy
accuracies = [accuracy_log_reg, accuracy_dt, accuracy_rf]
best_model_idx = accuracies.index(max(accuracies))

best_model = None
if best_model_idx == 0:
    best_model = "Logistic Regression"
elif best_model_idx == 1:
    best_model = "Decision Tree"
else:
    best_model = "Random Forest"

print("The Best Model based on Accuracy is:", best_model)
```
